# SCRAPPING DATA FROM WEBSSITE + PANDAS

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_South_Africa"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
colum_name = table.find_all('th')
```

```python
colum_name
```

```python
colum_table_names = [title.text.strip() for title in colum_name]

print(colum_table_names)
```

```python
df = pd.DataFrame(columns = colum_table_names)

df
```

```python
column_data = table.find_all('tr')
```

```python
# for loop starts from position 1 because position 0 has null values

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
```
